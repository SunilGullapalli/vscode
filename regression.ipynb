# Regression Example

```python
from sklearn.datasets import make_regression
```

```python
features = 20
X,Y = make_regression(n_samples=3000,n_features=features,n_informative=12,noise=50)
```

```python
cols = [f'x{i+1}' for i in range(features)]
cols
```

```python
import pandas as pd
X = pd.DataFrame(X,columns=cols)
X.head()
```

```python
Y = pd.DataFrame(Y,columns=['target'])
Y.head()
```

```python
X.info()
```

### Creating Sklearn pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
num_pipe = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
```

```python
X_pre  = pd.DataFrame(X_pre,columns=X.columns)
X_pre.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Building Ridge model with gridsearch cv

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
params = {'alpha':np.arange(0.1,30,0.1)}
params
```

```python
model = Ridge()
gscv = GridSearchCV(model,param_grid=params,cv=5,scoring='neg_mean_squared_error')
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_model = gscv.best_estimator_
best_model
```

## Predicting train and test values

```python
ypred_tr = best_model.predict(xtrain)
ypred_ts = best_model.predict(xtest)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
