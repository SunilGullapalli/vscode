```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=APLE&apikey=4RL5UIKRGMVTLL47'
r = requests.get(url)
data=r.json()
data
```

```python
ohlc= data['Time Series (Daily)']
```

```python
date =[]
open =[]
high= []
low=[]
close=[]
for key, value in ohlc.items():
    date.append(key)
    open.append(value['1. open'])
    high.append(value['2. high'])
    low.append(value['3. low'])
    close.append(value['4. close'])
```

```python
dct ={'Date':date, 'open':open, 'high':high, 'low':low, 'close':close}
```

```python
import pandas as pd
df =pd.DataFrame(dct)
df
```

```python
df.info()
```

```python
df.index=df['Date']
```

```python
df=df.drop(labels=['Date'],axis=1)
```

```python
df
```

```python
df1 =df[::-1]
df1
```

```python
df1['close']
```

```python
df1.index = pd.to_datetime(df1.index)
```

```python
df1['open']=df1['open'].astype(float)
df1['close']=df1['close'].astype(float)
df1['high']=df1['high'].astype(float)
df1['low']=df1['low'].astype(float)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,12))
plt.plot(df1['close'])
plt.show()
```
